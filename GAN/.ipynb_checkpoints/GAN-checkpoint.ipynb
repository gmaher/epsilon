```python
%matplotlib inline
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import os
import sys
sys.path.append(os.path.abspath("../.."))
import tensorflow_util.util as util_tf

sys.path.append(os.environ['DATASETS_PATH'])
from mnist.mnist import get_mnist
```

```python
#create training data
```

```python
datasets = get_mnist()
data, _ = datasets[0]
print data.shape
NUM_DATA = data.shape[0]

Npixels=28
Nchannels=1
data = data.reshape((NUM_DATA, Npixels, Npixels,1))
print data.shape
print data[0,:].shape
```

```python
#plot some random images
Num_images_to_plot = 5
for i in range(0,Num_images_to_plot):
    index = np.random.randint(NUM_DATA)
    
    plt.imshow(data[index,:,:,0])
    plt.show()
```

```python
#normalize data
```

```python
means = np.mean(data, axis=(0,1,2))
stdevs = np.std(data,axis=(0,1,2))

print means
print stdevs

data_norm = (data-means)/stdevs

print np.mean(data_norm, axis=(0,1,2))
print np.std(data_norm,axis=(0,1,2))
```

```python
#define network structure
NBATCH = 100
W_init = 0.001
NOISE_DIM = 100
G_Nunits1 = (Npixels-4)**2
```

```python
x = tf.placeholder(tf.float32, shape=(NBATCH,Npixels,Npixels,Nchannels))
z = tf.placeholder(tf.float32, shape=(NBATCH,NOISE_DIM))
mode = tf.placeholder(bool)
```

```python
#generator network
```

```python
G_hidden1, G_w1, G_b1 = util_tf.fc_weights(z, NOISE_DIM, G_Nunits1, std=W_init)
G_hidden1 = tf.nn.relu(G_hidden1)

G_hidden1 = tf.reshape(G_hidden1, [NBATCH, Npixels-4, Npixels-4, 1])

G_hidden1, G_gamma1, G_beta1 = util_tf.batch_norm(G_hidden1, mode, [1], axes=[0,1,2])

G_hidden2, G_w2, G_b2 = util_tf.deconv2d_3x3_weights(
G_hidden1,
outshape=[NBATCH, Npixels-2, Npixels-2, 32],
out_channels=32,
Nfilters=1,
std=W_init)

G_hidden2 = tf.nn.relu(G_hidden2)

G_hidden2, G_gamma2, G_beta2 = util_tf.batch_norm(G_hidden2, mode, [32], axes=[0,1,2])

G_out, G_w3, G_b3 = util_tf.deconv2d_3x3_weights(
G_hidden2,
outshape=[NBATCH, Npixels, Npixels, Nchannels],
out_channels=Nchannels,
Nfilters=32,
std=W_init)
```

```python
#Discriminator network
Nfilters = 32
Conv_Shape = 3
strides = [1,1,1,1]
p = 0.2
```

```python
D_w1 = tf.Variable(
    tf.random_normal(shape=(Conv_Shape,Conv_Shape,Nchannels,Nfilters),
                    stddev=W_init
                    )
)

D_b1 = tf.Variable(
    tf.zeros(shape=[Nfilters])
)

D_w2 = tf.Variable(
    tf.random_normal(shape=(Conv_Shape,Conv_Shape,Nfilters,Nfilters),
                    stddev=W_init
                    )
)

D_b2 = tf.Variable(
    tf.zeros(shape=[Nfilters])
)

D_w3 = tf.Variable(
    tf.random_normal(shape=(Nfilters*Npixels**2,1),
                    stddev=W_init
                    )
)

D_b3 = tf.Variable(
    tf.zeros(shape=(1,1))
)
```

```python
#pass real data through discriminator
D_real_out1 = util_tf.conv2d_3x3(x, D_w1, D_b1, padding="SAME")
D_real_out1 = tf.nn.relu(D_real_out1)

D_real_out1 = tf.nn.dropout(D_real_out1, keep_prob=p)

D_real_out2 = util_tf.conv2d_3x3(D_real_out1, D_w2, D_b2, padding="SAME")
D_real_out2 = tf.nn.relu(D_real_out2)
D_real_out1, D_r_gamma2, D_r_beta2\
= util_tf.batch_norm(D_real_out2, mode, [32], axes=[0,1,2])

D_real_out2 = tf.nn.dropout(D_real_out2, keep_prob=p)


D_real_out3 = tf.reshape(D_real_out2,[NBATCH,-1])
D_real_out3 = util_tf.fc(D_real_out3,D_w3, D_b3)
real_out = tf.nn.sigmoid(D_real_out3)
```

```python
#pass real data through discriminator
D_fake_out1 = util_tf.conv2d_3x3(G_out, D_w1, D_b1, padding="SAME")
D_fake_out1 = tf.nn.relu(D_fake_out1)
D_fake_out1 = tf.nn.dropout(D_fake_out1, keep_prob=p)

D_fake_out2 = util_tf.conv2d_3x3(D_fake_out1, D_w2, D_b2, padding="SAME")
D_fake_out2 = tf.nn.relu(D_fake_out2)
D_fake_out1, D_f_gamma2, D_f_beta2 = util_tf.batch_norm(D_fake_out2, mode, [32], axes=[0,1,2])
D_fake_out2 = tf.nn.dropout(D_fake_out2, keep_prob=p)


D_fake_out3 = tf.reshape(D_fake_out2,[NBATCH,-1])
D_fake_out3 = util_tf.fc(D_fake_out3,D_w3, D_b3)
fake_out = tf.nn.sigmoid(D_fake_out3)
```

```python
#loss functions
```

```python
var_reg = 0.0

D_loss = -tf.reduce_mean(tf.log(real_out)) -\
tf.reduce_mean(tf.log(1-fake_out))

#fake_var = tf.reduce_mean(tf.square(G_out-tf.reduce_mean(G_out)))

#var_loss = tf.square(fake_var-1)
var_loss = 0.0
G_loss1 = -tf.reduce_mean(tf.log(fake_out)) + var_reg*var_loss

G_loss2 = tf.reduce_mean(tf.log(1-fake_out)) + var_reg*var_loss
```

```python
#set up optimization
lrG = 1e-4
lrD = 1e-5
```

```python
optG = tf.train.AdamOptimizer(lrG)
optD = tf.train.AdamOptimizer(lrD)
D_train = optD.minimize(D_loss, var_list=[D_w1,D_b1,D_w2,D_b2,D_w3,D_b3,
                                         D_r_gamma2, D_r_beta2,
                                         D_f_gamma2, D_f_beta2])

G_train1 = optG.minimize(G_loss1, var_list=[G_w1,G_b1,G_w2,
                                            G_b2,G_w3,G_b3,
                                            G_gamma1, G_beta1,
                                           G_gamma2, G_beta2])

G_train2 = optG.minimize(G_loss2, var_list=[G_w1,G_b1,G_w2,G_b2,G_w3,G_b3,
                                           G_gamma1, G_beta1, G_gamma2,
                                           G_beta2])
```

```python
#set up session and initialize
init = tf.initialize_all_variables()
sess = tf.Session()
sess.run(init)
```

```python
#set up training loop
Niter1 = 1000
Niter2 = 1000
kd = 1
kg = 1
print_step = 100

def get_batch(xdata, nbatch):
    N = xdata.shape[0]
    inds = np.random.choice(N, size=nbatch, replace=False)
    xret = xdata[inds,:]

    return xret
```

```python
#some tests
x_batch = get_batch(data_norm, NBATCH)
z_batch = np.random.rand(NBATCH,NOISE_DIM)

gout = sess.run(G_out, feed_dict={z:z_batch, mode:True})
print gout.shape
print gout[0,0,:,:]
print sess.run(G_loss1, feed_dict={z:z_batch, mode:True})
print sess.run(G_loss2, feed_dict={z:z_batch, mode:True})

dout_real = sess.run(real_out, feed_dict={x:x_batch, mode:True})
dout_fake = sess.run(fake_out, feed_dict={z:z_batch, mode:True})
print dout_real.shape
print dout_fake.shape
print dout_real[0]
print dout_fake[0]
print sess.run(D_loss, feed_dict={x:x_batch,z:z_batch, mode:True})
```

```python
#training loop 1
for step in xrange(Niter1+1):

    for i in range(0,kd):    
        x_batch = get_batch(data_norm, NBATCH)
        z_batch = np.random.rand(NBATCH,NOISE_DIM)
        
        #sess.run(D_train,feed_dict={x:x_batch,z:z_batch, mode:True})
    
    for i in range(0,kg):
        z_batch = np.random.rand(NBATCH,NOISE_DIM)
        
        sess.run(G_train1, feed_dict={z:z_batch, mode:True})
        
    if step % print_step == 0:
        dl=sess.run(D_loss,feed_dict={x:x_batch,z:z_batch, mode:True})
        gl=sess.run(G_loss1,feed_dict={z:z_batch, mode:True})

        print "Discriminator loss: ", dl, ", Generator loss: ", gl
```

```python
#training loop 2
for step in xrange(Niter2+1):
    
    for i in range(0,kd):
        x_batch = get_batch(data_norm, NBATCH)
        z_batch = np.random.rand(NBATCH,NOISE_DIM)
    
        #sess.run(D_train,feed_dict={x:x_batch,z:z_batch, mode:True})
    
    for i in range(0,kg):
        z_batch = np.random.rand(NBATCH,NOISE_DIM)
        sess.run(G_train2, feed_dict={z:z_batch, mode:True})
        
    if step % print_step == 0:
        dl=sess.run(D_loss,feed_dict={x:x_batch,z:z_batch, mode:True})
        gl=sess.run(G_loss2,feed_dict={z:z_batch, mode:True})

        print "Discriminator loss: ", dl, ", Generator loss: ", gl
```
